```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo CSV
df = pd.read_csv('system_usage_log.csv', delimiter=',')

# Limpiar los nombres de las columnas para eliminar espacios adicionales
df.columns = df.columns.str.strip()

# Imprimir los nombres de las columnas y las primeras filas para verificar
print("Columnas disponibles:", df.columns)
print("Primeras filas del DataFrame:")
print(df.head())

# Asignar un índice secuencial
df.reset_index(drop=True, inplace=True)

# Crear el índice secuencial
df.index = range(len(df))

# Configurar la figura y los ejes para los gráficos
fig, axs = plt.subplots(3, 1, figsize=(10, 15), sharex=True)

# Graficar CPU Percent
if 'cpu_percent' in df.columns:
    axs[0].plot(df.index, df['cpu_percent'], marker='o', linestyle='-', color='r')
    axs[0].set_ylabel('CPU Percent')
    axs[0].set_title('CPU Usage Over Time')
else:
    axs[0].text(0.5, 0.5, "Columna 'cpu_percent' no encontrada", fontsize=12, ha='center')

# Graficar Memory Percent
if 'memory_percent' in df.columns:
    axs[1].plot(df.index, df['memory_percent'], marker='o', linestyle='-', color='g')
    axs[1].set_ylabel('Memory Percent')
    axs[1].set_title('Memory Usage Over Time')
else:
    axs[1].text(0.5, 0.5, "Columna 'memory_percent' no encontrada", fontsize=12, ha='center')

# Graficar Disk Usage Percent
if 'disk_usage_percent' in df.columns:
    axs[2].plot(df.index, df['disk_usage_percent'], marker='o', linestyle='-', color='b')
    axs[2].set_ylabel('Disk Usage Percent')
    axs[2].set_title('Disk Usage Over Time')
else:
    axs[2].text(0.5, 0.5, "Columna 'disk_usage_percent' no encontrada", fontsize=12, ha='center')

# Configurar el formato del índice en el eje X
axs[2].set_xlabel('Index')
plt.tight_layout()

# Mostrar los gráficos
plt.show()
```
